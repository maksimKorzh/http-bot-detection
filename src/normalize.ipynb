# Convert raw data into flattened dataset

```python
# packages
from tqdm import tqdm
import pandas as pd
import json, csv

# bot names
bot_names = [ 'bot', 'Bot', 'python-requests', 'Scrapy', 'curl', 'Wget']

# open raw data file stream
with open('./raw_data/access.log') as f:
    # loop over the entries
    for entry in tqdm(range(10365152)):
        # fetch next line from a file
        line = f.readline()

        # define format
        try:
            int(line.split('] "')[-1].split(' "-" "')[0].split(' ')[-2])
            status = line.split('] "')[-1].split(' "-" "')[0].split(' ')[-2]
            time = line.split('] "')[-1].split(' "-" "')[0].split(' ')[-1]
            agent = line.split(' "-" "')[-1].split('" "-"')[0]
        
        except:
            status = line.split('" ')[1].split(' "')[0].split(' ')[0]
            time = line.split('" ')[1].split(' "')[0].split(' ')[1]
            agent = line.split('" ')[-2].split(' "')[0][1:]
        
        # extract features
        try:
            features = {
                'ip': line.split(' - - ')[0],
                'date': line.split(' - - [')[-1].split(' ')[0],
                'zone': line.split(' - - [')[-1].split(']')[0].split(' ')[-1],
                'method': line.split('] "')[-1].split('"')[0].split(' ')[0],
                'url': line.split('] "')[-1].split('"')[0].split(' ')[1],
                'protocol': line.split('] "')[-1].split('"')[0].split(' ')[2],
                'status': status,
                'time': time,
                'agent': agent,
                'bot': 0
            }
        
        except: pass

        # label data
        for bot_name in bot_names:
            if bot_name in features['agent']:
                features['bot'] = 1
                break
        
        # if no user agent, most likely it's a bot
        if features['agent'] == '-': features['bot'] = 1

        try:
            # encode ip
            ip_encoded = [0] * 12    
            ip_raw = [ord(c) for c in features['ip'].replace('.', '')]
            for i in range(len(ip_raw)): ip_encoded[i] = ip_raw[i]

            # encode date
            date_raw = features['date'].replace('/', '').replace(':', '')
            date_encoded = [ord(c) for c in date_raw]

            # encode zone
            zone_raw = features['zone']
            zone_encoded = [ord(c) for c in zone_raw]

            # emncode method
            method_encoded = [0] * 4
            method_raw = features['method']
            for i in range(len(method_raw)): method_encoded[i] = ord(method_raw[i])

            # encode url
            url_encoded = [0] * 4000
            url_raw = features['url']
            for i in range(len(url_raw)): url_encoded[i] = ord(url_raw[i])
        
        
            # encode protocol
            protocol_raw = features['protocol'].replace('/', '').replace('.', '')
            protocol_encoded = [ord(c) for c in protocol_raw]

            # encode status
            status_raw = features['status']
            status_encoded = [ord(c) for c in status_raw]

            # encode time
            time_encoded = [0] * 8    
            time_raw = [ord(c) for c in features['time']]
            for i in range(len(time_raw)): time_encoded[i] = time_raw[i]

            # encode user agent
            agent_encoded = [0] * 300
            agent_raw = features['agent']
            for i in range(len(agent_raw)): agent_encoded[i] = ord(agent_raw[i])

            # encode all together
            data_encoded  = ip_encoded
            data_encoded += date_encoded
            data_encoded += zone_encoded
            data_encoded += method_encoded
            data_encoded += url_encoded
            data_encoded += protocol_encoded
            data_encoded += status_encoded
            data_encoded += time_encoded
            data_encoded += agent_encoded

            # flattened row
            encoded_row = { 'Data': data_encoded, 'Label': features['bot']}

            # append line to flattened dataset
            with open('./train_data/data.csv', 'a') as d:
                csv.DictWriter(d, encoded_row.keys()).writerow(encoded_row)
        
        except: pass
```
